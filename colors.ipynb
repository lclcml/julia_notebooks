# Basic Types¶

The most basic library for working with colors in Julia is `ColorTypes.jl`. More information about it can be found at https://github.com/JuliaGraphics/ColorTypes.jl.  The most basic type it defines is the abstract type `Colorant` which allows definition of concrete types of colors like `RGB{T}`.  The `T` parameter allows us to make a color out of different Types such as Floats or Ints, demonstrated below in boxes 2 and 3. Other color types include `HSV{T}` (Hue, Saturation, Value), `RGBA{T}` (Red, Green, Blue, Alpha), `LCHuv{T}` (Luminance, Chroma, Hue), and `Gray{T}`.  ![](https://github.com/JuliaGraphics/ColorTypes.jl/raw/master/images/types.png) There are many more colortypes available based on [CIE](https://en.wikipedia.org/wiki/International_Commission_on_Illumination) colorspaces if you need them.

Since most people work with RGB, let's create a red color in RGB...

```julia
# add Packages if they aren't already in your system
import Pkg; Pkg.add("ColorTypes"); Pkg.add("Colors")
```

```julia
using ColorTypes
```

```julia
# define red
# this should return the line: RGB{Float64}(1.0,0.0,0.0)
#  since not using Colors yet
RGB(1.0,0.0,0.0)
```

```julia
RGB(1,0,0)
```

# ColorMaps

What if we want a range of colors? We want a `colormap`.

`colormap()` creates an `Array{ColorTypes}` with specified N=number of colors and mid=midpoint

    colormap(cname::String [N::Int=100; mid=0.5, logscale=false, kvs...])
    
These maps use a particular hue and generate a path through LCH space (Luminance, Chroma, Hue) that is smooth, without abrupt discontinuities.  A primary source for this algorithm is Wijffelaars, M., Vliegen, R., van Wijk, J., van der Linden, E-J. (2008). [Generating Color Palettes using Intuitive Parameters](https://pdfs.semanticscholar.org/0977/49c130c1cf35b8b9c236916de3b0455ffce0.pdf). More background can be found in Wijffelaars, M., & Wijk, J. (2008). [Synthesis of color palettes](http://alexandria.tue.nl/extra1/afstversl/wsk-i/wijffelaars2008.pdf) (Doctoral dissertation, Master’s thesis. Supervisors Wijk, J. van, and Vliegen).


## Sequential palettes

Sequential maps use a single hue and are created with the `sequential_palette()` function. 

    sequential_palette(h, N::Int64; c, s, b, w, d, wcolor, dcolor, logscale) 

* `N` (number of colors in palette) defaults to 100 as above. 
* `c` is contrast from 0-1
* `s` is saturation from 0-1
* `b` is brightness from 0-1

if you want the rest, see timholy's great documentation at [colors.jl](https://github.com/JuliaGraphics/Colors.jl)

Let's look at what some of the different hues look like if we travel around the circle in degrees.

```julia
# These should render as a spectrum of colors from light to dark in iJulia. Otherwise, you see the Array of colors below.
sequential_palette(30.0)
```

## Diverging palette

Two hues can be combined in a divergent palette. In this case, the first two arguments are the hues at the two ends of the palette.  A lot of control can be specified through the `w` argument (strength of the middle color from 0-1) and `wcolor` (the middle color).  There is one predefined named diverging palette called `RdBu` (from Red to Blue).

```julia
colormap("RdBu")
```

```julia
Colors.colormaps_diverging
```

```julia
diverging_palette(30.0, 240.0, w=0)
```

```julia
# when we make the middle green, the divergent nature is more apparent
diverging_palette(30.0, 240.0, w=1, wcolor=RGB(0,1,0))
```

```julia
# this can be softened by changing w
diverging_palette(30.0, 240.0, w=0.5, wcolor=RGB(0,1,0))
```

Now I'm going to aim for a rainbow, by flipping this around and altering the saturation, contrast, and brightness parameters. These adjustments will actually make the map less efficient for discrimination from a psychophysics point of view, but they are color patterns that you will frequently see in many statistical plots. 

```julia
diverging_palette(270, 1, w=1, wcolor=RGB(1,1,0))
```

```julia
diverging_palette(240, 1, w=1, wcolor=RGB(1,1,0), s=1, b=1, c=1)
```

```julia
diverging_palette(300, 1, w=1, wcolor=RGB(1,1,0), s=1, b=1, c=1)
```
